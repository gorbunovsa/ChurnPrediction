# 1.1 Анализ корреляций
Исследуем коэффициенты корреляции между числовыми признаками.

```python
corr_mat = df.corr()
corr_size = corr_mat.shape[0]
corr_mat.shape
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(corr_mat, cmap = 'coolwarm', annot = False)
plt.show()
```

Если обучать модель на данных, содержащих несколько сильно скоррелированных признаков, то модель может переобучиться на информации, которую описывают данные признаки. Поэтому из двух сильно скоррелированных признаков необходимо оставить один (приоритетно удалять будем признак содержащий больше пропусков).

```python
columns_null_perc = df.isnull().sum(axis = 0)/df.shape[0]
columns_null_perc
```

```python
corr_threshold = 0.9
drop_columns = set(['Customer_ID'])
for j in range(1,corr_size):
  for i in range(j):
    if abs(corr_mat.iloc[i,j]) >= corr_threshold:
      row = corr_mat.index[i]
      col = corr_mat.columns[j]
      if columns_null_perc[col] > columns_null_perc[row]:
        drop_columns.add(col)
      else:
        drop_columns.add(row)
list(drop_columns)
```

```python
df_nocorr = df.drop(columns = drop_columns)
df_nocorr.shape
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(df_nocorr.corr(), cmap = 'coolwarm', annot = False)
plt.show()
```

# 1.2 Анализ и заполнение пропусков в данных

```python
plt.figure(figsize = (30,5))
sns.heatmap(df_nocorr.isnull())
plt.show()
```

```python
def columns_info(df, columns, print_items = True):
  for col in columns:
    print(col)
    if print_items:
      print(df[col].unique(), df[col].nunique())
    else:
      print('Number of unique values: ', df[col].nunique())
    print('Number of NaNs: ', df[col].isnull().sum())
    print('')
```

```python
cat_columns = set(df_nocorr.columns[df_nocorr.nunique() <= 20])
columns_info(df_nocorr, cat_columns)
```

```python
str_columns = set(df_nocorr.columns[df_nocorr.dtypes == 'object'])
columns_info(df_nocorr, str_columns)
```

```python
null_columns = set(df_nocorr.columns[df_nocorr.isnull().any(axis = 0)])
df_nocorr[null_columns].isnull().sum(axis = 0)/df_nocorr.shape[0]
```

```python
df_nonull = df_nocorr.copy()
df_nonull.head()
```

Пропуски в категориальных и порядковых переменных будем заполнять модой.

```python
ord_null_columns = (cat_columns & null_columns) - str_columns
columns_info(df_nocorr, ord_null_columns)
```

```python
for col in ord_null_columns:
  df_nonull[col].fillna(df_nonull[col].mode()[0], inplace = True)

df_nonull[ord_null_columns].isnull().sum().sum()
```

Пропуски в строковых переменных kid{число}_{число}, подразумевающих ответ на вопрос от детях соответствующего возраста и имеющих значение 'U', будем заполнять значением 'U' (unknown). Даже если эта догадка неверна, значение 'U' является модой в этих столбцах. Аналогично поступим с переменной hnd_webcap, имеющей значение 'UNKW'.
  
Пропуски в переменных с большим количеством пропусков (dwllsize, infobase, HHstatin, dwlltype, ownrent) также заполним значением 'U', тем самым создавая новую группу клиентов, которые не предоставили соответствующей информации (ранее значения 'U' среди значений этих переменных не было).

Пропуски в категориальных и порядковых строковых переменных, имеющих малое число пропусков, заполним модой.

```python
str_null_columns = (cat_columns & null_columns) - ord_null_columns
columns_info(df_nocorr, str_null_columns)
```

```python
for col in ['kid6_10', 'kid3_5', 'kid16_17', 'kid0_2', 'kid11_15', 'dwllsize', 'infobase', 'HHstatin', 'dwlltype', 'ownrent']:
  df_nonull[col].fillna('U', inplace = True)

df_nonull['hnd_webcap'].fillna('UNKW', inplace = True)

for col in ['prizm_social_one', 'creditcd', 'dualband', 'refurb_new', 'marital', 'ethnic', 'area']:
  df_nonull[col].fillna(df_nonull[col].mode()[0], inplace = True)

df_nonull[str_null_columns].isnull().sum().sum()
```

Пропуски в числовых переменных можно заполнить средним либо медианой. Как показывают диграммы размаха ниже, переменные содержат много выбросов, поэтому для заполнения пропусков будет использована медиана. Более того, переменная phones, на самом деле является порядковой, поэтому пропуски в ней будут заполнены модой.

```python
cont_null_columns = null_columns - ord_null_columns - str_null_columns
columns_info(df_nocorr, cont_null_columns, print_items = False)
```

```python
plt.figure(figsize = (15,15))
for i, col in enumerate(cont_null_columns, start = 1):
  plt.subplot(3, 3, i)
  sns.boxplot(data = df_nocorr[col], orient = 'v')
  plt.title(col)
plt.show()
```

```python
df_nonull['phones'].fillna(df_nonull['phones'].mode()[0], inplace = True)
cat_columns.add('phones')
for col in cont_null_columns - set(['phones']):
  df_nonull[col].fillna(df_nonull[col].median(), inplace = True)
```

Итак, все пропуски заполнены.

# 1.3 Кодирование нечисловых переменных
Для кодирования категориальных признаков, содержащих малое количество уникальных значений (не более четырех), будет использовано one-hot кодирование.

Для кодирования категориальных признаков с большим количеством уникальных значений будет использовано частотное кодирование.

Для кодирования порядковых признаков будет использовано порядковое кодирование.

```python
columns_info(df_nonull, str_columns)
```

```python
def freq_encoder(vec):
  freqs = vec.value_counts()/len(vec)
  return vec.apply(lambda x : freqs[x])
```

```python
one_hot_enc_columns = ['kid11_15', 'kid6_10', 'kid0_2', 'kid3_5', 'kid16_17', 
                       'infobase', 'new_cell', 'hnd_webcap', 'asl_flag', 
                       'creditcd', 'dualband', 'refurb_new', 'ownrent', 'dwlltype']
freq_enc_columns = ['HHstatin', 'marital', 'ethnic', 'area', 'prizm_social_one']
ord_enc_columns = ['dwllsize', 'crclscod']

churn = pd.get_dummies(df_nonull, columns = one_hot_enc_columns, drop_first=True)
for col in freq_enc_columns:
  churn[col] = freq_encoder(churn[col])
for col in ord_enc_columns:
  churn[col] = pd.factorize(churn[col], sort=True)[0]
churn
```

# 1.4 Нормализация данных, разделение на тренировочную и тестовую выборки

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(churn, test_size = .25 , random_state = 123)
X_train = train.drop(['churn'], axis = 1)
X_test = test.drop(['churn'], axis = 1)
Y_train = train['churn']
Y_test = test['churn']
```

Выделим переменные, которые изначально были бинарными.

```python
not_dummy_columns = list(set(df_nonull.columns) - set(one_hot_enc_columns))
binary_columns = [not_dummy_columns[i] for i in np.where(df_nonull[not_dummy_columns].nunique() == 2)[0]]
binary_columns
```

Обучим scaler на тренировочных данных (получим средние и стандартные отклонения), применим scaler к тренировочной и тестовой выборкам. 

```python
from sklearn.preprocessing import StandardScaler
columns_to_scale = list(set(df_nonull.columns) - set(one_hot_enc_columns) - set(binary_columns))
scaler = StandardScaler()
X_train[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])
X_test[columns_to_scale] = scaler.transform(X_test[columns_to_scale])
```

# **2. Обучение моделей и прогнозирование**


Для классификации были выбраны следующие модели:
*   Логистическая регрессия;
*   Метод k-ближайших соседей;
*   Дерево решений;
*   Случайный лес;
*   Градиентный бустинг.

Параметры моделей выбираются поиском по сетке на кросс-валидации.







```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

classifiers = {
    'Logistic Regression': LogisticRegression(solver = 'liblinear'),
    'kNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state = 0),
    'Random Forest': RandomForestClassifier(random_state = 0, bootstrap = True),
    'XGBoost': XGBClassifier(random_state = 0)
}

grids = {
    'kNN': {'n_neighbors': [15, 25, 35]},
    'Decision Tree': {'max_depth': [3, 9, 15]},
    'Random Forest': {'n_estimators': [10, 100, 300]},
    'XGBoost': {'max_depth': [3, 9, 15],
                'learning_rate': [0.1, 0.2, 0.3]}
}

def classifier_fit_predict(name, grid = None):
  if name in grids:
    clf = GridSearchCV(estimator = classifiers[name],
                       param_grid = grids[name],
                       scoring = 'accuracy',
                       cv = 5)
  else:
    clf = classifiers[name]
  clf.fit(X_train, Y_train)
  pred = clf.predict(X_test)
  prob = clf.predict_proba(X_test)[:,1]
  cm = confusion_matrix(Y_test, pred)
  fpr, tpr, _ = roc_curve(Y_test, prob)
  fig, axes = plt.subplots(1, 2, figsize=(18, 8))
  fig.suptitle(name, fontsize = 30)
  sns.heatmap(cm, annot = True, fmt = 'd', xticklabels = ['No Churn', 'Churn'], 
              yticklabels = ['No Churn', 'Churn'], cmap = 'viridis', ax = axes[0])
  axes[0].set_title('Confusion Matrix', fontsize = 20)
  axes[0].set_xlabel('Predicted Label')
  axes[0].set_ylabel('True Label')
  axes[1].plot(fpr, tpr, color = 'royalblue', linewidth = 3)
  axes[1].plot([0, 1], [0, 1], color = 'orchid', linewidth = 3)
  axes[1].set_title('ROC Curve', fontsize = 20)
  axes[1].set_xlabel('False Positive Rate')
  axes[1].set_ylabel('True Positive Rate')
  plt.show()
  metrics = [accuracy_score(Y_test, pred), precision_score(Y_test, pred), 
             recall_score(Y_test, pred), roc_auc_score(Y_test, prob)]
  print('Accuracy  = {:.4f}'.format(metrics[0]))
  print('Precision = {:.4f}'.format(metrics[1]))
  print('Recall    = {:.4f}'.format(metrics[2]))
  print('AUC       = {:.4f}'.format(metrics[3]))
  if name in grids:
    print('Best parameters are', clf.best_params_)
  return metrics
```

```python
df_metrics = pd.DataFrame(data = None, 
                          index = ['Logistic Regression', 
                                   'kNN', 
                                   'Decision Tree', 
                                   'Random Forest', 
                                   'XGBoost'],
                          columns = ['Accuracy',
                                     'Precision',
                                     'Recall',
                                     'AUC'])
```

```python
df_metrics.loc['Logistic Regression'] = classifier_fit_predict('Logistic Regression')
```

```python
df_metrics.loc['kNN'] = classifier_fit_predict('kNN')
```

```python
df_metrics.loc['Decision Tree'] = classifier_fit_predict('Decision Tree')
```

```python
df_metrics.loc['Random Forest'] = classifier_fit_predict('Random Forest')
```

```python
df_metrics.loc['XGBoost'] = classifier_fit_predict('XGBoost')
```

# **3. Сравнение моделей**

По всем метрикам лучшей оказалась модель градинтного бустинга, однако случайный лес не сильно уступает XGB. При более тщательной подборке параметров обе эти модели могут показать достойные результаты.

В этих данных классы переменной churn сбалансированы, поэтому стоит обращать внимание на метрику accuracy. Однако более важно не упускать клиентов, которые собираются уйти. Нам необходимо, чтобы алгоритм находил как можно больше объектов положительного класса из всех объектов положительного класса (recall), и при этом, чтобы как можно больше объектов, названных алгоритмом положительными, действительно были положительными (precision). Большое значение FP, означало бы, что мы тратим много ресурсов на клиентов, которые и так не собираются уходить, а большое значение FN - что мы упускаем клиентов, которые собираются уйти.
